```python
def plot_correlation_matrix(corr):
    # Generate a mask for the upper triangle
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(11, 9))

    # Generate a custom diverging colormap
    cmap = sns.diverging_palette(240, 10, n=9)

    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .8})
    plt.show()
```

```python
corr =  numeric_features.corr()
plot_correlation_matrix(corr)
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_features = categorical_features.replace('\<', '', regex=True)
first=0
for col in categorical_features.columns:  
    lb_make = LabelEncoder()
    new_feature = lb_make.fit_transform(categorical_features[col].astype(str))
    df = pd.DataFrame(new_feature, columns=[col])

    if first == 0:
        encoded_categorical_features = df
        first=1
    else:
        encoded_categorical_features = pd.concat([encoded_categorical_features, df], axis=1)
```

```python
corr=encoded_categorical_features.corr()
plot_correlation_matrix(corr)
```
